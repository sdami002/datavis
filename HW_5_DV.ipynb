```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a pandas DataFrame
data = pd.read_csv('https://www2.census.gov/programs-surveys/popest/datasets/2010-2019/national/totals/nst-est2019-alldata.csv', encoding='latin-1')

# Extract the RBIRTH2019, RDEATH2019, and RNETMIG2019 columns
rates = data[['RBIRTH2019', 'RDEATH2019', 'RNETMIG2019']]

# Create the boxplot
plt.boxplot(rates.values, labels=['Birth Rates', 'Death Rates', 'Migration Rates'])

# Set the plot title and axis labels
plt.title('Distribution of Birth, Death, and Migration Rates in the US (2019)')
plt.xlabel('Rate Type')
plt.ylabel('Rate')

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.distributions.empirical_distribution import ECDF

# Load the dataset into a pandas DataFrame
data = pd.read_csv('https://www2.census.gov/programs-surveys/popest/datasets/2010-2019/national/totals/nst-est2019-alldata.csv', encoding='latin-1')

# Extract the RBIRTH2019 column
birth_rates = data['RNETMIG2019']

# Create the histogram
plt.hist(birth_rates, bins=15)

# Set the plot title and axis labels
plt.title('Distribution of Birth Rates in the US (2019)')
plt.xlabel('Birth Rate')
plt.ylabel('Frequency')

# Show the plot
plt.show()

# Create the eCDF
ecdf = ECDF(birth_rates)
x = np.sort(birth_rates)
y = ecdf(x)

# Create the plot
plt.step(x, y)

# Set the plot title and axis labels
plt.title('Empirical CDF of Birth Rates in the US (2019)')
plt.xlabel('Birth Rate')
plt.ylabel('Cumulative Probability')

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset into a pandas DataFrame
data = pd.read_csv('https://www2.census.gov/programs-surveys/popest/datasets/2010-2019/national/totals/nst-est2019-alldata.csv', encoding='latin-1')

# Extract the RNETMIG2019 column
migration_rates = data['RNETMIG2019']

# Create the histogram with separate frames for each bar
plt.hist(migration_rates, bins=15, edgecolor='black')

# Calculate the mean of the migration rates
mean = np.mean(migration_rates)

# Add the mean line to the histogram
plt.axvline(x=mean, color='red')

# Set the plot title and axis labels
plt.title('Distribution of Net Migration Rates in the US (2019)')
plt.xlabel('Net Migration Rate')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a pandas DataFrame
data = pd.read_csv('https://www2.census.gov/programs-surveys/popest/datasets/2010-2019/national/totals/nst-est2019-alldata.csv', encoding='latin-1')

# Extract the data for birth, death, and net migration rates
birth_rates = data['RBIRTH2019']
death_rates = data['RDEATH2019']
migration_rates = data['RNETMIG2019']

# Create the line chart
plt.plot(birth_rates, label='Birth Rates')
plt.plot(death_rates, label='Death Rates')
plt.plot(migration_rates, label='Net Migration Rates')

# Set the plot title and axis labels
plt.title('Trends in Birth, Death, and Net Migration Rates in the US (2010-2019)')
plt.xlabel('Year')
plt.ylabel('Rate')

# Add a legend to the plot
plt.legend()

# Show the plot
plt.show()
```
