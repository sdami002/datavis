# **Part 1 code**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset from a CSV file
df = pd.read_csv("https://data.virginia.gov/resource/vsrk-d6hx.csv")
df = df.drop_duplicates(subset=['week_ending_date', 'vaccination_status'], keep='first')
# Calculate the weekly rate per 100,000 people for each vaccination status
df['Rate_per_100k'] = (df['infections'] / df['population_denominator']) * 100000
# remove the "T00:00:00.000" from the report_date column
df['report_date'] = df['report_date'].str.slice(stop=10)

# remove the "T00:00:00.000" from the week_ending_date column
df['week_ending_date'] = df['week_ending_date'].str.slice(stop=10)

# Pivot the data to have vaccination status as columns and week as index
df_pivot = df.pivot(index='week_ending_date', columns='vaccination_status', values='Rate_per_100k')

# Create a line chart
df_pivot.plot(kind='line', marker='o')

# Set chart properties
plt.xlabel('week_ending_date')
plt.ylabel('Rate_per_100k')
plt.title('Weekly COVID-19 Cases in Virginia by Vaccination Status')
plt.legend(title='Vaccination Status')
plt.xticks(rotation=45)

# Show the chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://data.virginia.gov/resource/28k2-x2rj.csv")

df = df[~df['health_district'].isin(['Out of State', 'Not Reported'])]


df['vaccine_manufacturer'] = df['vaccine_manufacturer'].str.replace('Pfizer.*', 'Pfizer')


df = df[df['dose_number'] == 1]


pfizer_prop = df[df['vaccine_manufacturer'] == 'Pfizer']['dose_number'].groupby(df['health_district']).sum() / df['dose_number'].groupby(df['health_district']).sum()


pfizer_prop.plot(kind='bar', figsize=(12,6))
plt.title("Proportion of 1st Doses that were Pfizer by Health District")
plt.xlabel("Health District")
plt.ylabel("Proportion of Doses time 1")
plt.show()

```
